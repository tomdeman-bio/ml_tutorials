```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
a = 0.5
b = 1.0

# x from 0 to 10
x = 30 * np.random.random(1000)
print('x')
print(x)
```

```python
y = a * x + b + np.random.normal(size=x.shape)
print('y')
print(y)
```

```python
reg = LinearRegression()
reg.fit(x[:, None], y)
print('model')
print(reg)
```

```python
print('Estimated a:', reg.coef_)
print('Estimated b:', reg.intercept_)
```

```python
x_new = np.linspace(0, 30, 10)
y_new = reg.predict(x_new[:, None])
print('x_new')
print(x_new)
```

```python
print('y_new')
print(y_new)
```

```python
ax = plt.axes()
ax.scatter(x, y, color='green')
ax.scatter(x_new, y_new, color='blue')
ax.plot(x_new, y_new, color='black')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axis('tight')
plt.show()
```
